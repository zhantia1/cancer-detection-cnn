```python
def display_random_images_grid(label_data, train_data_path, num_images=10, grid_cols=3):
    # Get the total number of images
    total_images = len(label_data)
    
    random_indices = random.sample(range(total_images), num_images)
    
    # Calculate the number of rows needed for the grid
    grid_rows = -(-num_images // grid_cols)  # Ceiling division
    
    # Create a grid of subplots
    fig, axes = plt.subplots(grid_rows, grid_cols, figsize=(10, 10))
    
    # Flatten the axes if there's only one row
    if grid_rows == 1:
        axes = [axes]
    
    # Display images corresponding to the random indices
    for i, ax in zip(random_indices, axes.flatten()):
        image = cv2.imread(train_data_path + label_data['id'][i] + '.tif')
        height, width, channels = image.shape
        print("Image size:", height, "x", width, "pixels")
        print("Number of channels:", channels)
        ax.imshow(image, resample=True, cmap='gray')
        ax.set_title(label_data['label'][i])
        ax.axis('off')
    
    # Adjust layout to prevent overlapping
    plt.tight_layout()
    plt.show()

display_random_images_grid(label_data, train_data_path, num_images=12, grid_cols=3)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

image_height=96
image_width=96
num_channels=3

# Define the CNN architecture
model = Sequential([
    # Convolutional layer with 32 filters, each 3x3 in size, using ReLU activation
    Conv2D(32, (3, 3), activation='relu', input_shape=(image_height, image_width, num_channels)),
    # Max pooling layer with a pool size of 2x2
    MaxPooling2D((2, 2)),
    
    # Convolutional layer with 64 filters, each 3x3 in size, using ReLU activation
    Conv2D(64, (3, 3), activation='relu'),
    # Max pooling layer with a pool size of 2x2
    MaxPooling2D((2, 2)),
    
    # Convolutional layer with 128 filters, each 3x3 in size, using ReLU activation
    Conv2D(128, (3, 3), activation='relu'),
    # Max pooling layer with a pool size of 2x2
    MaxPooling2D((2, 2)),
    
    # Flatten layer to convert 3D feature maps to 1D feature vectors
    Flatten(),
    
    # Fully connected (dense) layer with 128 neurons, using ReLU activation
    Dense(128, activation='relu'),
    
    # Output layer with softmax activation for multiclass classification
    Dense(1, activation='sigmoid')
])

model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```
